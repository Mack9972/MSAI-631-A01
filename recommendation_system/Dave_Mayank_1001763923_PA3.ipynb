### Task 2: Run apriory.py and Evaluate Results 

In this task, I executed and print apriory results by making use only the apriory.py. 

Execute apriori algorithm "3" times for different combinations of support and confidence.

Displaying results of apriory for 'dataset.csv'.

Explaining my reasoning 




```python
#Reading dataset.csv and converting into list.
with open('dataset.csv', 'r') as read_csv_file:
    csv_data = csv.reader(read_csv_file)
    transactions = list(csv_data)
```

```python
print ('# # # # # # # # #  Code for Task 2, Case:1 # # # # # # # # #') 
items, rules = runApriori(transactions, 0.1, 0.70)
print ('Case 1 (minimum support=0.1 and minimum confidence=0.50)')
print ('Case 1 Reasoning: \n\nWhen minimum support and minimum confidence are 0.1 and 0.5, mineral water is bought about 23.8% of total transactions while eggs are the second most frequent bought item at 18% followed by spaghetti, French fires, chocolate, green tea and milk. Also there are no rules because there are no items with min confidance above 0.5 ')
print ('Case 1 Output:')
printResults(items,rules)
```

```python
print ('# # # # # # # # #  Code for Task 2, Case:2 # # # # # # # # #') 
items, rules = runApriori(transactions, 0.50, 0.50)
print ('Case 2 (minimum support=0.50 and minimum confidence=0.50)') 
print ('Case 2 Reasoning: \n\nWhen minimum support and confidence is 50%, result item and rules are none because we do not have items with min support and confidence above 0.5.')
print ('Case 2 Output:') 
printResults(items,rules)
```

```python
print ('# # # # # # # # #  Code for Task 2, Case:3 # # # # # # # # #') 
items, rules = runApriori(transactions, 0.01, 0.4)
print ('Case 3 (minimum support=0.01 and minimum confidence=0.3)')
print ('Case 3 Reasoning:\n\n When minimum support and minimum confidence are .01 and .3, mineral water is bought about 23.8% of total transactions and it can also be seen that mineral water is bought by about half 50.7% of the people who buys ground beef and eggs. Similarly mineral water is bought alongside with milk and ground beef about 50.3%.' )
print ('Case 3 Output:')
printResults(items,rules)
```

### Task 3: Run FPgrowth and Evaluate Results 
By executing this task, I was able to to execute and print FPgrowth results using mlxtend.

Executed apriori algorithm "3" times for different combinations of support and confidence. Showed the results of FPgrowth for 'dataset.csv'. 

Included my reasoning to explain the result outcome

 

```python
#Reading dataset.csv and converting into list.
with open('dataset.csv', 'r') as read_csv_file:
    csv_data = csv.reader(read_csv_file)
    dataset_data = list(csv_data)
    
    
tran_encoder = TransactionEncoder()
transaction = tran_encoder.fit(dataset_data).transform(dataset_data)
df = pd.DataFrame(transaction, columns=tran_encoder.columns_)
df
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:1 # # # # # # # # #') 
print ('Case 1 (minimum support=0.1)')
print ('Case 1 Reasoning:When minimum support and minimum confidence are 0.1 and 0.5, mineral water is bought about 23.8% of total transactions while eggs are the second most frequent bought item at 18% followed by spaghetti, French fires, chocolate, green tea and milk. Also there are no rules because there are no items with min confidance above 0.5. One thing to note that the output in ')
print ('Case 1 Output:')
frequent_itemsets = fpgrowth(df, min_support=0.1,use_colnames = True)
frequent_itemsets


```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules = rules[['antecedents','consequents','support','confidence']]
rules
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:2 # # # # # # # # #') 
print ('Case 1 (minimum support=0.5)')
print ('Case 1 Reasoning:When minimum support is 0.03 and confidence is 50%, we get items that are bought frequently but we get no rules as we do not have items with min support and confidence above 0.5.')
print ('Case 1 Output:')
frequent_itemsets =fpgrowth(df, min_support=0.03 , use_colnames = True)
frequent_itemsets 
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules = rules[['antecedents','consequents','support','confidence']]
rules
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:3 # # # # # # # # #') 
print ('Case 1 (minimum support=0.01)')
print ('Case 1 Reasoning:When minimum support and minimum confidence are .01 and .3, mineral water is bought about 23.8% of total transactions and it can also be seen that mineral water is bought by about half 50.7% of the people who buys ground beef and eggs. Similarly mineral water is bought alongside with milk and ground beef about 50.3%.')
print ('Case 1 Output:')
frequent_itemsets =fpgrowth(df, min_support=0.01,use_colnames = True)
frequent_itemsets 
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
rules = rules[['antecedents','consequents','support','confidence']]
rules
```
